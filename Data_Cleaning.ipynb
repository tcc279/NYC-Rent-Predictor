```python
#data cleaning
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_set2_no_dupes.csv')
```

```python
df = df[df.Bathrooms != '-- baths']
df = df[df.Bathrooms != '988.5 baths']
```

```python
df['Bathrooms'] = df['Bathrooms'].map(lambda x: x.lstrip('+-').rstrip('baths'))
```

```python
#studios are recorded as zero bedrooms, needs to be reinterpreted as 1 bedroom and also a studio
Studio = []
for i in df['Bedrooms']:
    print(i)
    if i == int(0):
        print('is studtio')
        Studio.append(int(1))
    else:
        Studio.append(int(0))
```

```python
df['Studio'] = pd.Series(Studio)
```
