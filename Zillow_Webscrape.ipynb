```python
links_list = []
full_list = []


#Get apartment listings from each URL
def link_list_generator(url_list):
    for url in url_list:
        user_agent = {'User-agent': 'Mozilla/5.0'}
        response  = requests.get(url, headers = user_agent)
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        links = [link for link in soup.find_all('a') if 'routable' in str(link)]
        for a in links:
            links_list.append(a['href'])
        for i in links_list:
            j = "https://www.zillow.com" + i
            full_list.append(j)
        time.sleep(.5+2*random.random())
```

```python
#many of the links fromthe Zillow webpage are building links - not for the actual apartment
#in order to get the individual apartment listings, the building links need to be separated from the full_list  

def building_seperator(url_list):
    for url in url_list:
        y = re.search(r"\w/b/\w+", url)
        if y is None:
            continue
        else:
            building_link = y.string
            building_list.append(url)
            url_list.remove(url)
```

```python
#this function will get the apartment listings from the building links page

def building_filter(building_list):
    for url in building_list:
        user_agent = {'User-agent': 'Mozilla/5.0'}
        response  = requests.get(url, headers = user_agent)
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        if soup.find(id='units-panel-for-rent') is None:
            print ("%s has NoneType ..." %(url))
            continue
        else:
            add_url = [i['href'] for i in soup.find(id='units-panel-for-rent').find_all('a')]
            url_list.append(add_url)
            time.sleep(.5+2*random.random())
```

```python
#create lists for features to be appended to
address_list = []
bedrooms_list = []
city_list = []
hood_list = []
price_list = []
county_list = []
bath_list = []
laundry_list = []
heat_list = []
cooling_list = []
pets_list = []
park_list = []
```

```python
#function to scrape info off page/s
def info_scrape(url_list):
    for url in url_list:
        print(url)
        user_agent = {'User-agent': 'Mozilla/5.0'}
        response  = requests.get(url, headers = user_agent)
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        #address = soup.find_all('h1', {'class':'notranslate' })[0].text
        #address_list.append(address)
        google_ad_config = soup.find_all('div', {'google-ad-config' })[1].text
        #convert json to dictionary
        datastore = json.loads(google_ad_config)
        #get metrics dictionary
        metrics = datastore['targets']
        
        #address
        address = metrics['aamgnrc1']
        address_list.append(address)
        
        
        #number of bedrooms
        bedrooms = metrics['bd']
        bedrooms_list.append(bedrooms)
        
        #city
        city = metrics['city']
        city_list.append('city')
        
        #hood
        hood = metrics['hood']
        hood_list.append(hood)
        
        #rental price
        price = metrics['price']
        price_list.append(price)
        
        #county
        county = metrics['cnty']
        county_list.append(county)
        
        #number of baths
        bath = soup.find_all('span', {'class':'addr_bbs' })[1].text
        bath_list.append(bath)
        
        #laundry
        laundry = soup.find_all('div', {'class':'hdp-fact-ataglance-value' })[1].text
        laundry_list.append(laundry)
        
        #heat
        heat = soup.find_all('div', {'class':'hdp-fact-ataglance-value' })[2].text
        heat_list.append(heat)
        
        #cooling
        cooling = soup.find_all('div', {'class':'hdp-fact-ataglance-value' })[3].text
        cooling_list.append(cooling)
        
        #pets
        pets = soup.find_all('div', {'class':'hdp-fact-ataglance-value' })[4].text
        pets_list.append(pets)
        
        #park
        park = soup.find_all('div', {'class':'hdp-fact-ataglance-value' })[5].text
        park_list.append(park)
        
        time.sleep(.5+2*random.random())
```
